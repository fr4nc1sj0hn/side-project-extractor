```python
import pandas as pd
```

```python
columnPlaceholder = [
    'Event Type',
    'Timestamp',
    'Log Message',
    'Category',
    'User',
    'Thread ID',
    'Error Code',
    'Application',
    'Machine'
]
columnPlaceholder = columnPlaceholder + ['col' + str(i) for i in range(60)]
```

```python
path = "C:\\Users\\Francis John Picaso\\Repositories\\Teo\\logs\\GeneralLog1350185-220110.log"
df = pd.read_csv(path, sep='\t', lineterminator='\n',names=columnPlaceholder)
df
```

```python
categories = [
    "MainUI.CAO",
    "MainUI.DataTransfer_ILink",
    "MainUI.DBB",
    "MainUI.DiskAdmin",
    "MainUI.General_System",
    "RR.BLIF.Infrastructure"
]
correct_format = df[df['Category'].isin(categories)]
merged = df.merge(correct_format, how='left',left_index=True, right_index=True, indicator=True)
noncorrect_format = merged[merged["_merge"] == "left_only"]
noncorrect_format
```

```python
correct_format
```

```python
#process incorrect format
corrected = pd.DataFrame()
EventTypes = []
Timestamps = []
LogMessages = []
Categories = []
indexes = []
category_column_markers = []
for index, row in noncorrect_format.iterrows():
    colid = 0
    message = ""
    indexes.append(index)
    EventTypes.append(row['Event Type_x'])
    Timestamps.append(row['Timestamp_x'])
    for col, value in row.items():
        if value in categories:
            Category = value
            category_column_markers.append(colid)
            break
        if colid > 1:
            message = message + ('' if pd.isna(value) else value)
        colid = colid + 1
    LogMessages.append(message)
    Categories.append(Category)
    
corrected["Event Type_x"] = EventTypes
corrected["Timestamp_x"] = Timestamps
corrected["Log Message_x"] = LogMessages
corrected["Category_x"] = Categories
corrected.index = indexes
corrected
```
